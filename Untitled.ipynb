```python
grid = [[0, 0, 0, 0, 0, 0, 6, 8, 0],
        [0, 0, 0, 0, 7, 3, 0, 0, 9],
        [3, 0, 9, 0, 0, 0, 0, 4, 5],
        [4, 9, 0, 0, 0, 0, 0, 0, 0],
        [8, 0, 3, 0, 5, 0, 9, 0, 2],
        [0, 0, 0, 0, 0, 0, 0, 3, 6],
        [9, 6, 0, 0, 0, 0, 3, 0, 8],
        [7, 0, 0, 6, 8, 0, 0, 0, 0],
        [0, 2, 8, 0, 0, 0, 0, 0, 0]]

valid_move_count = 0
solve_count = 0

def is_valid_move(grid, row, col, num):
    global valid_move_count
    valid_move_count += 1
    print("valid", valid_move_count)
    valid_flag = True
    # is there something in row
    for x in range(9):
        if grid[row][x]==num:
            valid_flag = False
            return valid_flag
    # is there something in column
    for x in range(9):
        if grid[x][col]==num:
            valid_flag = False
            return valid_flag
    # is there something in block        
    if row%2==0:
        row_start = row
        if row+2 > 8:
            row_start = row-2
    else:
        row_start = row-1
    row_end = row_start+2
    
    if col%2==0:
        col_start = col
        if col+2 > 8:
            col_start = col-2
    else:
        col_start = col-1
    col_end = col_start+2

    for i in range(row_start, row_end+1):
        for j in range(col_start, col_end+1):
            if grid[i][j] == num:
                return False
    return valid_flag


def solve(grid, row, col):
    global solve_count
    solve_count += 1
    print("solve", solve_count)
    if col == 9:
        if row == 8:
            return True
        row+=1
        col=0
    if grid[row][col] > 0:
        return solve(grid, row, col+1)
    for num in range(1,10):
        if is_valid_move(grid, row, col, num):
            grid[row][col] = num
            
            if solve(grid, row, col+1):
                return True
            
        grid[row][col] = 0
    return False

for i in range(100):
    if solve(grid, 0, 0):
        for i in range(9):
            for i in range(9):
                print(grid[i][j], end=" ")
            print()
    else:
        print("No Solution Found")
```
